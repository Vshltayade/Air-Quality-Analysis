```python
df_cleaned = df.dropna(subset=['pollutant_avg']).copy()
station_avg = df_cleaned.groupby(['latitude', 'longitude']).agg(
    mean_pollutant_avg=('pollutant_avg', 'mean')
).reset_index()
plt.figure(figsize=(8, 10))
scatter = plt.scatter(
    station_avg['longitude'],
    station_avg['latitude'],
    c=station_avg['mean_pollutant_avg'],
    cmap='Reds', # Use a 'Reds' colormap to signify increasing pollution
    s=50, # Set a fixed size for all points
    alpha=0.8,
    edgecolors='k',
    linewidths=0.5
)
plt.xlim(68, 98) # Longitude range for India
plt.ylim(8, 38)  # Latitude range for India
cbar = plt.colorbar(scatter, label=r'Average Pollutant Concentration')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Air Quality Stations in India (Colored by Average Pollutant Level)', fontsize=14)
plt.grid(alpha=0.5)
plt.gca().set_aspect(1.3)

plt.tight_layout()
```

```python
df_cleaned = df.dropna(subset=['pollutant_avg']).copy()
maharashtra_data = df_cleaned[df_cleaned['state'] == 'Maharashtra'].copy()
if maharashtra_data.empty:
    print("Error: No data found for Maharashtra.")
else:
    station_avg = maharashtra_data.groupby(['latitude', 'longitude']).agg(
        mean_overall_avg=('pollutant_avg', 'mean')
    ).reset_index()
    min_val = station_avg['mean_overall_avg'].min()
    max_val = station_avg['mean_overall_avg'].max()
    if max_val == min_val:
        scaled_sizes = 50
    else:
      scaled_sizes = 50 + 400 * (station_avg['mean_overall_avg'] - min_val) / (max_val - min_val)

    plt.figure(figsize=(8, 10))

    scatter = plt.scatter(
        station_avg['longitude'],
        station_avg['latitude'],
        c=station_avg['mean_overall_avg'],
        cmap='plasma', # Use the high-contrast colormap
        s=scaled_sizes, # Size is scaled by concentration
        alpha=0.8,
        edgecolors='k',
        linewidths=0.5
    )
    plt.xlim(72, 82)
    plt.ylim(15, 23)

    cbar = plt.colorbar(scatter, label=r'Overall Average Pollutant Concentration')

    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title('Overall Average Pollutant Concentration Across Stations in Maharashtra', fontsize=14)
    plt.grid(alpha=0.5)

    plt.gca().set_aspect(1.3)

    plt.tight_layout()
```

```python
df_cleaned = df.dropna(subset=['pollutant_avg']).copy()
df_cleaned
pollutant_pivot = df_cleaned.pivot_table(
    index=['state', 'city', 'station', 'latitude', 'longitude'],
    columns='pollutant_id',
    values='pollutant_avg',
    aggfunc='mean'
)
pollutant_pivot
pollutant_cols = pollutant_pivot.columns
pollutant_cols
pollutant_pivot_filled = pollutant_pivot.fillna(pollutant_pivot.mean())
pollutant_pivot_filled
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(pollutant_pivot_filled)
scaled_df = pd.DataFrame(scaled_features, columns=pollutant_cols, index=pollutant_pivot_filled.index)
scaled_df
print("Shape of the processed and scaled data for clustering:", scaled_df.shape)
print("\nFirst 5 rows of scaled data:")
print(scaled_df.head().to_markdown(numalign="left", stralign="left"))
```

```python
wcss = []
for i in range(1,11):
  kmeeans = KMeans(n_clusters=i, init='k-means++',random_state=41)
  kmeeans.fit(scaled_features)
  wcss.append(kmeeans.inertia_)
plt.plot(range(1,11),wcss)
```

```python
# Perform K-Means Clustering with K=4
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans.fit(scaled_features)
# 1. Get Cluster Centers
cluster_centers_scaled = kmeans.cluster_centers_
# 2. Inverse Transform to original scale using the live scaler object
# Formula: original = scaled * scale + mean
cluster_centers_original = cluster_centers_scaled * scaler.scale_ + scaler.mean_
# Convert to DataFrame for readability
cluster_centers_df = pd.DataFrame(cluster_centers_original, columns=pollutant_cols)
cluster_centers_df.index.name = 'Cluster'
# 3. Assign Cluster Labels to the original data
pollutant_pivot_filled['Cluster'] = kmeans.labels_
cluster_counts = pollutant_pivot_filled['Cluster'].value_counts().sort_index()
# 4. Characterize and rename clusters based on PM2.5 and PM10 levels (the most critical pollutants)
# (Sorting cluster centers by PM2.5 average to ensure consistent naming)
sorted_clusters = cluster_centers_df.sort_values(by='PM2.5').index

cluster_names = {
    sorted_clusters[0]: "1 - Very Low Pollution",
    sorted_clusters[1]: "2 - Low Pollution",
    sorted_clusters[2]: "3 - Moderate Pollution",
    sorted_clusters[3]: "4 - High Pollution"
}

# Apply names to the cluster centers and counts
cluster_centers_df['Profile'] = cluster_centers_df.index.map(cluster_names)
cluster_counts_named = cluster_counts.rename(index=cluster_names)
cluster_centers_df = cluster_centers_df.set_index('Profile').round(2)

print("--- Cluster Pollution Profiles (Average Pollutant Concentration) ---")
print(cluster_centers_df.to_markdown(floatfmt=".1f"))

print("\n--- Station Count by Cluster ---")
print(cluster_counts_named.to_frame(name='Station Count').to_markdown())

# Prepare data for plotting the geographical distribution of clusters (Next step)
pollutant_pivot_filled = pollutant_pivot_filled.reset_index()
pollutant_pivot_filled['Cluster Name'] = pollutant_pivot_filled['Cluster'].map(cluster_names)
```

```python
# Load the clustered data
clustered_df = pollutant_pivot_filled
# Define a consistent order and color map for the clusters
cluster_order = ['4 - High Pollution', '3 - Moderate Pollution', '2 - Low Pollution', '1 - Very Low Pollution']
palette = {
    '4 - High Pollution': 'darkred',
    '3 - Moderate Pollution': 'red',
    '2 - Low Pollution': 'orange',
    '1 - Very Low Pollution': 'green'
}
# Create the geographical scatter plot
plt.figure(figsize=(8, 10))

# Use seaborn scatterplot for easy categorical coloring
sns.scatterplot(
    data=clustered_df,
    x='longitude',
    y='latitude',
    hue='Cluster Name',
    hue_order=cluster_order,
    palette=palette,
    s=70, # Size of the points
    alpha=0.8,
    edgecolor='k',
    linewidth=0.5
)
# Set map limits to roughly cover India
plt.xlim(68, 98) # Longitude range for India
plt.ylim(8, 38)  # Latitude range for India

# Add labels and title
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Distribution of Air Quality Clusters (K=4)', fontsize=14)
plt.grid(alpha=0.5)

# Move legend outside the plot
plt.legend(title='Pollution Profile', bbox_to_anchor=(1.05, 1), loc='upper left')

# Ensure the aspect ratio is correct for a map
plt.gca().set_aspect(1.3)

plt.tight_layout()
```
